you ar given N sticks of negligible thickness and the ith stick has A[i]
you have to form a rectangle by choosing any four sticks .
Find the maximum possible areaof the rectanle.
*note:
a rectangle is a figure having opposite sides are equal
a square is also a rectangle
write the function rectangle which has following parameters
return value:integer denoting maximum possible arae
1.Number of sticks(integer)
2.Integer array denoting length of sticks
Algorithm:
remove thosde sticks which do not have pair
ex:
input(7,[2,8,2,8,8,3,3,4])#[2,8,3] [16,6,24]
 (a,b)
output:24

```python
#n=no.of sticks and side_lengths=length of them (changed 'len' to avoid built-in clash)
def rectangle(n, side_lengths):
    arr = []
    for i in side_lengths:
        
        
        if side_lengths.count(i) >= 2 and i not in arr:
            arr.append(i)
            # Fix: If a number appears 4 or more times, add it twice so it can form a square
            if side_lengths.count(i) >= 4:
                arr.append(i)

    max_area = 0
    # Fix: Unindented these loops so they run AFTER finding all available side pairs
    for i in range(len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i] * arr[j] > max_area:
                max_area = arr[i] * arr[j]
    return max_area

print(rectangle(8, [1, 2, 2, 3, 3, 4, 4]))
```
